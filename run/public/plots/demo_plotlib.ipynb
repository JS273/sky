# Plotlib demo

Notebook to demonstrate the available plot types

### Line Plot

```python
from sky.plots.plotlib import LinePlot, Plotter
import numpy as np

x = np.linspace(0.0,10,100)
y = np.sin(x)

lin_plt = LinePlot(x,y)

my_plotter = Plotter()
_ = my_plotter.plot(lin_plt)
```

### Contour Plot

```python
from sky.plots.plotlib import ContourPlot, Plotter
from sky.plots.datastructures import Grid
import numpy as np

# Create Contour plot
domain = Grid(np.linspace(0.0, 10, 100), np.linspace(0.0, 5, 50))
y = np.sin(domain.nodes[:,0]) * np.cos(domain.nodes[:,1])
cont_plt = ContourPlot(domain, y)
# cont_plt.draw_xticks = False
# cont_plt.draw_yticks = True
# cont_plt.cbar = True
# cont_plt.cbar_ticks = [-0.5, 0.0, 0.5]

# Plotter
my_plotter = Plotter()
_ = my_plotter.plot(cont_plt)
```

### Histogram plot

```python
from sky.plots.plotlib import HistPlot, Plotter
import numpy as np

samples = np.random.normal(0.0, 4.0, 1000)
hist_plt = HistPlot(y = samples)

# Plotter
my_plotter = Plotter()
_ = my_plotter.plot(hist_plt)
```

### Plot all plots in subplot and save

```python
from sky.plots.plotlib import *
from sky.filemanager import *

# Setup
tag = "Subplots"
result_folder = create_result_folder(tag = tag, max_res_folders = 7, max_daily_folders= 7)

# Create plotter
plotter = Plotter(save_path = result_folder, open_saved_plot=False)
_ = plotter.plot(lin_plt, hist_plt, cont_plt, filename = f'{tag}')
```
